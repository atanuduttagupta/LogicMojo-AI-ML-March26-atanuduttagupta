Q1. Division with Exception Handling

Write a program that takes two numbers and safely performs division.
Handle division by zero using try-except.

````python
def divide(num1, num2):
    result = None
    try:
        result = num1/num2
    except ZeroDivisionError as e:
        print(e)
    except Exception as e:
        print('other exception', e)
    else:
        print('no exception')
    finally:
        print('all the time')
    return result

result = divide(1,'a')
print(result)
````

Q3. Multiple Exception Handling

Write a program that:

Takes user input
Converts it to integer
Divides 100 by that number

Handle:

ValueError
ZeroDivisionError

````python
def take_input():
    text = input('Enter a number:')
    try:
        num = int(text)
        result = 100/num
        print(result)
    except ValueError as e:
        print(f"Can not convert {text} to integer")
    except ZeroDivisionError as e:
        print(f"Can not divide 100 by 0")        
    except Exception as e:
        print('General Exception', e)
take_input()
````

## 2. Text Cleaning Pipeline

You are given:

```python
texts = [
    "Hello!!! How are you???",
    "This is    AI     class.",
    "Python@@@is###great"
]
```

Task:

1. Write a function clean_text(text):
2. Remove special characters
3. Remove extra spaces
4. Convert text to lowercase
5. Apply the function to all elements using loop or map
6. Return cleaned list

---

````python
import re

texts = [
    "Hello!!! How are you???",
    "This is    AI     class.",
    "Python@@@is###great"
]


def clean_text(text):
    text = re.sub(r'[^A-Za-z0-9 ]', '', text)
    text = re.sub(r'\s+', ' ', text)
    return text.lower().strip()

new_text = list(map(clean_text, texts))

print(new_text)
    
````

## 3. Fraud Detection Rule Engine

You are given:

```python
transactions = [
    {"id": 101, "amount": 5000, "location": "India"},
    {"id": 102, "amount": 200000, "location": "Unknown"},
    {"id": 103, "amount": 1000, "location": "India"}
]
```

Task:

1. Write a function is_fraud(transaction)
2. If amount is greater than 100000 or location is "Unknown" mark as fraud
3. Otherwise mark as safe
4. Use lambda with filter or map
5. Return only fraud transactions


````python
transactions = [
    {"id": 101, "amount": 5000000, "location": "India"},
    {"id": 102, "amount": 200000, "location": "Unknown"},
    {"id": 103, "amount": 1000, "location": "India"}
]

def is_fraud(transaction):
    return transaction["amount"] > 100000 or transaction["location"] == "Unknown"

result = list(filter(lambda t:is_fraud(t), transactions))

print(result)
````

## 4. Feature Engineering

You are given:

```python
data = [
    {"salary": 50000, "experience": 2},
    {"salary": 100000, "experience": 5}
]
```

Task:

1. Write a function feature_engineer(record)
2. Add a new key salary_per_exp
3. salary_per_exp is salary divided by experience
4. Handle division by zero using conditional logic
5. Return updated dataset

---

````python
data = [
    {"salary": 50000, "experience": 2},
    {"salary": 100000, "experience": 5}
]

def feature_engineer(record):
    try:
        record["salary_per_exp"] = record["salary"]/record["experience"]
    except Exception as e:
        record["salary_per_exp"] = None
    return record


new_data = list(map(lambda t:feature_engineer(t), data))
print(new_data)
````

## 8. Recursive Decision Tree Simulation
Task:

Write a recursive function decision_tree(score)
If score is greater than 80 return "Approved"
If score is between 50 and 80 call the function again with score plus 10
If score is less than 50 return "Rejected"

````python
def decision_tree(score):
    if score > 80:
        return "Approved"
    elif score < 50:
        return "Rejected"
    else:
        return decision_tree(score + 10)

print(decision_tree(51))
````
