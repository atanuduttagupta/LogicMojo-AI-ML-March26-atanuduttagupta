```python
from sklearn.datasets import load_iris
import numpy as np
import pandas as pd

iris = load_iris()

x = iris['data']
x.shape

mean = np.mean(x, axis = 0)
median = np.mean(x, axis = 0)
std_dev = np.std(x, axis = 0)
variance = np.var(x, axis = 0)

print('\nMean: ', mean)
print('Median: ', median)
print('Std Deviation: ', std_dev)
print('Variance: ', variance)

max_var_index = np.argmax(variance)

feature_names = iris.feature_names

print('\nFeature with maximum variance: ', feature_names[max_var_index])
print('Value of maximum variance: ', variance[max_var_index])


feature = x[:, 0]
print('original shape: ', feature.shape)

reshaped_feature= feature.reshape(-1,1)
print('reshaped shape: ', reshaped_feature.shape)

#print(feature)
#print(reshaped_feature)
```

Q11. End-to-End ML Preprocessing Pipeline (Titanic)

Context: Prepare dataset for ML model training

Tasks:

1. Load dataset
2. Handle missing values
3. Encode categorical variables (sex, embarked)
4. Normalize numerical features
5. Convert to NumPy array

Output:

Final feature matrix ready for model training

```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.preprocessing import StandardScaler

# 1. Load dataset
df = sns.load_dataset('titanic')

# 2. Select relevant columns (important step)
df = df[['survived', 'pclass', 'sex', 'age', 'sibsp', 'parch', 'fare', 'embarked']]

# 3. Handle missing values

# Numerical → fill with median
df['age'].fillna(df['age'].median(), inplace=True)
df['fare'].fillna(df['fare'].median(), inplace=True)

# Categorical → fill with mode
df['embarked'].fillna(df['embarked'].mode()[0], inplace=True)

# 4. Encode categorical variables

# sex → binary encoding
df['sex'] = df['sex'].map({'male': 0, 'female': 1})

# embarked → one-hot encoding
df = pd.get_dummies(df, columns=['embarked'], drop_first=True)

# 5. Separate features and target
X = df.drop('survived', axis=1)
y = df['survived']

# 6. Normalize numerical features
scaler = StandardScaler()

num_cols = ['age', 'fare', 'sibsp', 'parch']
X[num_cols] = scaler.fit_transform(X[num_cols])

# 7. Convert to NumPy array
X_final = X.values
y_final = y.values

# 8. Output
print("Feature Matrix Shape:", X_final.shape)
print("Target Shape:", y_final.shape)

print("\nSample Features:\n", X_final[:5])
print("\nSample Target:\n", y_final[:5])
```

Q12. Feature Selection using Statistical Understanding

Context: Not all features improve model performance.

Dataset: Iris

Tasks:

1. Compute correlation matrix
2. Identify highly correlated features
3. Drop redundant features
4. Explain:
Impact on overfitting and model performance

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris

# Load dataset
iris = load_iris()

df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Correlation matrix
corr_matrix = df.corr()

print(corr_matrix)

# Threshold for high correlation
threshold = 0.9

high_corr_pairs = []

for i in range(len(corr_matrix.columns)):
    for j in range(i):
        if abs(corr_matrix.iloc[i, j]) > threshold:
            col1 = corr_matrix.columns[i]
            col2 = corr_matrix.columns[j]
            high_corr_pairs.append((col1, col2))

print("Highly correlated feature pairs:")
print(high_corr_pairs)

# Drop one feature from correlated pair
df_reduced = df.drop(columns=['petal width (cm)'])

print("Original shape:", df.shape)
print("Reduced shape:", df_reduced.shape)

# highly correlated feature means duplicate information. Model sees same signal twice
```

Q13. Custom Feature Transformation using apply()

Context: Real ML requires custom logic transformation.

Dataset: Titanic

Tasks:

1. Create age_group:
Child / Adult / Senior
2. Use .apply()
3. Compute survival rate per group
4. Interpret:
Which segment has highest survival likelihood

```python
import pandas as pd
import seaborn as sns

df = sns.load_dataset('titanic')

# Keep relevant columns
df = df[['age', 'survived']]


def age_category(age):
    if pd.isnull(age):
        return 'Unknown'
    elif age < 18:
        return 'Child'
    elif age < 60:
        return 'Adult'
    else:
        return 'Senior'

df['age_group'] = df['age'].apply(age_category)

print(df[['age', 'age_group']].head())

survival_by_group = df.groupby('age_group')['survived'].mean()

print(survival_by_group)

# children has highest survival likelihood
```

Q14. Error Analysis Simulation (Model Evaluation Thinking)

Context: Understanding prediction errors is key in ML.

Tasks:

1. Generate 1000 random error values using NumPy
2. Compute:
Mean
Standard deviation
3. Identify outliers:
Using mean ± 2*std
4. Explain:
What high variance means in model performance

```python
import numpy as np

np.random.seed(42)

# Generate 1000 random errors (normal distribution)
errors = np.random.normal(loc=0, scale=1, size=1000)

mean_error = np.mean(errors)
std_error = np.std(errors)

print("Mean Error:", mean_error)
print("Standard Deviation:", std_error)


lower_bound = mean_error - 2 * std_error
upper_bound = mean_error + 2 * std_error

outliers = errors[(errors < lower_bound) | (errors > upper_bound)]

print("Outliers count:", len(outliers))
print("Sample outliers:", outliers[:10])

# high variance means model predictions are inconsistent, sometimes accurrate, sometimes inconsistent
```

Q15. Business + ML Insight Problem (Tips Dataset)

Context: Convert data analysis into decision-making

Tasks:

1. Create:
tip_percentage
2. Group by:
day and time
3. Identify:
Highest revenue segment
4. Provide:
Recommendation for business strategy
How ML model can use this insight

```python
import pandas as pd
import seaborn as sns

tips = sns.load_dataset('tips')

# Create tip percentage
tips['tip_percentage'] = (tips['tip'] / tips['total_bill']) * 100

print(tips.head())

grouped = tips.groupby(['day', 'time']).agg({
    'total_bill': 'sum',
    'tip': 'sum',
    'tip_percentage': 'mean'
}).reset_index()

print(grouped)

# Highest revenue based on total_bill
top_segment = grouped.loc[grouped['total_bill'].idxmax()]

print("Highest Revenue Segment:\n", top_segment)
```
