## Task 3 :Identify primary and foreign keys 

```python
def identify_keys(df, primary_key, foreign_key):
    df_name = df.attrs.get('name', '')
    display(HTML(f"<b>{'*'*30} Print Keys for Dataframe {df_name} {'*'*30}</b>"))  
    total_rows = len(df)
    unique_rows = df.drop_duplicates().shape[0]
    print('Total Rows : ', total_rows)
    print('Unique Rows : ', unique_rows) 
    print('Duplicates: ', 'N' if total_rows == unique_rows else 'Y')
    # check keys
    if primary_key:
        df = df.drop_duplicates() 
        if isinstance(primary_key, list) and df[primary_key].duplicated().sum() == 0:
            print('Primary Key: ', primary_key)
        elif isinstance(primary_key, str) and df[primary_key].nunique() == total_rows:
            print('Primary Key: ', primary_key)                
        else:            
            print('Primary Key: None')            

    if foreign_key and isinstance(foreign_key, dict):
        for key, values in dict.items(foreign_key):
            print(f"Foreign Key: {key} to Dataset: {values}")
    else:            
        print('Foreign Key: None')    

identify_keys(df_location , ['geolocation_zip_code_prefix', 'geolocation_lat', 'geolocation_lng'], '') 
identify_keys(df_customers , 'customer_id', {'customer_zip_code_prefix': 'location, Column- geolocation_zip_code_prefix'} )  
identify_keys(df_sellers , 'seller_id', {'seller_zip_code_prefix': 'location, Column- geolocation_zip_code_prefix'})  
identify_keys(df_cat_trans , 'product_category_name', '')  
identify_keys(df_products , 'product_id', {'product_category_name': 'category_translation, Column- product_category_name'})  
identify_keys(df_orders , 'order_id', {'customer_id': 'customers, Column- customer_id'})  
identify_keys(df_order_item , ['order_id', 'order_item_id'], {'order_id': 'orders, Column- order_id', 'product_id': 'products, Column- product_id', 'seller_id': 'sellers, Column- seller_id'})  
identify_keys(df_payments , ['order_id', 'payment_sequential', 'payment_type'], {'order_id': 'orders, Column- order_id'})  
identify_keys(df_reviews , ['review_id', 'order_id'], {'order_id': 'orders, Column- order_id'}) 
```

## Task 1: Handle missing values appropriately 

```python
# check missing values
def check_missing_values(df):
    df_name = df.attrs.get('name', '')
    display(HTML(f"<b>{'*'*30} Check Missing Values for Dataframe {df_name} {'*'*30}</b>"))  
    print('** Check for Nan------>')
    print(df.isna().sum())
    check_empty = (df.select_dtypes(include='object')
                    .apply(lambda x: x.str.strip())
                    .eq('')
                    .sum())
    print('\n** Check for Empty String------>')    
    print(check_empty)
```

```python
check_missing_values(df_customers) # no missing value
check_missing_values(df_location) # no missing value
check_missing_values(df_sellers) # no missing value
check_missing_values(df_products) # missing value
check_missing_values(df_cat_trans) # no missing value

check_missing_values(df_orders) # missing value
check_missing_values(df_order_item) # no missing value
check_missing_values(df_payments) # no missing value
check_missing_values(df_reviews) # missing value
```

## Handle Missing Values in Products

```python
# drop data where very small number of missing values
df_products = df_products.dropna(subset=['product_weight_g', 'product_length_cm', 'product_width_cm', 'product_height_cm'])

# product_category_name is important categoricl value and we will fill with unknown
df_products['product_category_name'] = df_products['product_category_name'].fillna('unknown')

# product_photos_qty is number, we can fill with 0
df_products['product_photos_qty'] = df_products['product_photos_qty'].fillna(0)

# product_name_lenght and product_description_lenght are derived column, for unknown categories we fill up with 0, 
# if it was random, we could have used mean/median
df_products['product_name_lenght'] = df_products['product_name_lenght'].fillna(0)
df_products['product_description_lenght'] = df_products['product_description_lenght'].fillna(0)
```

```python
check_missing_values(df_products) # missing value
```

## Handle Missing Values in Orders

```python
print('Length: ', df_orders.shape[0])
check_missing_values(df_orders) # missing value
# we cannot drop this columns as missing value means order not approved/ not delivered
# we can create flags new features
df_orders['is_approved'] = df_orders['order_approved_at'].notnull()
df_orders['is_shipped'] = df_orders['order_delivered_carrier_date'].notnull()
df_orders['is_delivered'] = df_orders['order_delivered_customer_date'].notnull()
```

## Handle Missing Values in Reviews

```python
# reviews are optional, so we have lots of missing values
# replace empty strings
df_reviews['review_comment_title'] = df_reviews['review_comment_title'].str.strip().replace('', None)
df_reviews['review_comment_message'] = df_reviews['review_comment_message'].str.strip().replace('', None)

# we cannot put mode for random reviews, we create new feature
df_reviews['has_comments'] = df_reviews['review_comment_title'].notnull()
df_reviews['has_message'] = df_reviews['review_comment_message'].notnull()
```

```python
check_missing_values(df_reviews) # missing value
```

## Task 2: 	Remove duplicate records 

```python
# row level duplicates
def duplicate_check(df):
    df_name = df.attrs.get('name', '')
    display(HTML(f"<b>{'*'*3} Check Duplicate Values for Dataframe {df_name} {'*'*3}</b>"))
    has_duplicate = ('Yes' if df.shape[0] != df.drop_duplicates().shape[0]
                     else 'No')
    print('Duplicate: ', has_duplicate)
    print("\n")

duplicate_check(df_customers)
duplicate_check(df_location)
duplicate_check(df_sellers)
duplicate_check(df_products)
duplicate_check(df_cat_trans)

duplicate_check(df_orders)
duplicate_check(df_order_item)
duplicate_check(df_payments)
duplicate_check(df_reviews)
```

### df_location  has duplicates at row level, we have to remove it

```python
import unicodedata

# check for key level duplicates and clean
df_location = df_location.drop_duplicates()
duplicate_check(df_location)

print(list(df_location['geolocation_city'].unique())[0:100])
# has some characters which are unwanted, need to clean text
def clean_text(text):
    if text == '':
        return text
    text = text.lower().strip()
    text = unicodedata.normalize('NFKD', text)
    text = text.encode('ascii', 'ignore').decode('utf-8')
    text = text.replace('-', ' ')
    text = ' '.join(text.split())
    return text

df_location['geolocation_city'] = df_location['geolocation_city'].apply(clean_text)
```

## Task 3: Convert date columns to datetime format 
required for time analysis (monthly trends, delays, etc.) and Avoids string comparison error

```python
def convert_date(df, cols):
    for col in cols:
        df[col] = pd.to_datetime(df[col], errors = 'coerce')

convert_date(df_orders, ['order_purchase_timestamp','order_approved_at','order_delivered_carrier_date',
                                     'order_delivered_customer_date', 'order_estimated_delivery_date'])
convert_date(df_order_item, ['shipping_limit_date'])
convert_date(df_reviews, ['review_creation_date', 'review_answer_timestamp'])
```

```python
def standardize_column(df):
    df.columns = (
        df.columns.str.strip()
        .str.lower()
        .str.replace(' ', '_')
        .str.replace('-', '_')
    )
    return df
```

```python
df_customers = standardize_column(df_customers)
df_location = standardize_column(df_location)
df_sellers = standardize_column(df_sellers)
df_products = standardize_column(df_products)
df_cat_trans = standardize_column(df_cat_trans)
df_orders = standardize_column(df_orders)
df_order_item = standardize_column(df_order_item)
df_payments = standardize_column(df_payments)
df_reviews = standardize_column(df_reviews)
```
